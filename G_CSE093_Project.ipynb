## Data Cleaning

### Checking For Missing Values

```python
df.isnull().sum()
```

**No Missing Values in the Dataset**

### Dropping irrelevant columns

```python
df.drop(['EmployeeCount','EmployeeNumber','Over18','StandardHours'],axis=1,inplace=True)
```

```python
df.head()
```

### Scaling Numerical Columns

```python
from sklearn.preprocessing import StandardScaler
numerical_cols = df.select_dtypes(include=['number','float']).columns
numerical_cols = numerical_cols.drop('Attrition')
scaler = StandardScaler()

from sklearn.pipeline import Pipeline


numerical_transformer = Pipeline(steps=[
    ('scaler', scaler)
])
```

### Encoding Categorical Columns

```python
from sklearn.preprocessing import OneHotEncoder

from sklearn.compose import ColumnTransformer

categorical_transformer = Pipeline(steps=[
    ('onehot', OneHotEncoder(handle_unknown='ignore', sparse_output=False))
])

cat_cols = df.select_dtypes(exclude=['number','float']).columns


preprocessor = ColumnTransformer(
    transformers=[
        ('num', numerical_transformer, numerical_cols),
        ('cat_ohe', categorical_transformer, cat_cols)
    ],
    remainder='passthrough'
)
```

```python
df.head()
```

## Data Splitting and Handling




```python
from sklearn.model_selection import train_test_split

X_clf = df.drop('Attrition', axis=1)
y_clf = df['Attrition']

X_train_clf, X_test_clf, y_train_clf, y_test_clf = train_test_split(X_clf, y_clf, test_size=0.2, random_state=38)
```

```python
X_train_preprocessed_clf = preprocessor.fit_transform(X_train_clf)

X_test_preprocessed_clf = preprocessor.transform(X_test_clf)
```

```python
y_train_clf.value_counts()
```

**The Dataset is imbalanced**

### Using SMOTE to balance the dataset

```python
from imblearn.over_sampling import SMOTE
sm = SMOTE(random_state=42,sampling_strategy=
           0.5)
X_train_bal, y_train_bal = sm.fit_resample(X_train_preprocessed_clf, y_train_clf)
```

```python
y_train_bal.value_counts()
```

**Balanced train dataset**

# Part 1 - Model Training for Classification

## Logistic Regression on balanced dataset

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, roc_auc_score, classification_report,f1_score
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.ensemble import VotingClassifier
```

```python
logreg_bal = LogisticRegression(
    class_weight='balanced',
    max_iter=1000,
    random_state=42
)
logreg_bal.fit(X_train_bal, y_train_bal)

y_pred = logreg_bal.predict(X_test_preprocessed_clf)

accuracy = accuracy_score(y_test_clf, y_pred)
print(f"Accuracy of Logistic Regression: {accuracy}")
y_proba = logreg_bal.predict_proba(X_test_preprocessed_clf)[:, 1]
roc_auc = roc_auc_score(y_test_clf, y_proba)
print(f"ROC AUC of Logistic Regression: {roc_auc}")

print(classification_report(y_test_clf, y_pred))

f1 = f1_score(y_test_clf, y_pred,average='weighted')
print(f"F1 Score: {f1}")
```

## Best Model

```python
from imblearn.pipeline import Pipeline as ImbPipeline

clf_lr = LogisticRegression(
    penalty='l1',
    solver='saga',
    C=79.32047656808551,
    l1_ratio=0.4667628932479799,
    random_state=42,
    max_iter=1000,
    class_weight='balanced'
)

clf_rf = RandomForestClassifier(
    n_estimators=216,
    max_depth=21,
    min_samples_split=10,
    min_samples_leaf=5,
    max_features='log2',
    random_state=42,
    n_jobs=-1,
    class_weight='balanced'
)

clf_xgb = XGBClassifier(
    n_estimators=299,
    max_depth=4,
    learning_rate=0.08536189862866832,
    subsample=0.7218455076693483,
    colsample_bytree=0.9795542149013333,
    objective='binary:logistic',
    use_label_encoder=False,
    eval_metric='logloss',
    random_state=42,
    n_jobs=-1,
    scale_pos_weight=(y_train_clf.value_counts()[0] / y_train_clf.value_counts()[1]),
)


voting_clf = VotingClassifier(
    estimators=[
        ('lr', clf_lr),
        ('rf', clf_rf),
        ('xgb', clf_xgb)
    ],
    voting='soft',
    weights=[2, 1, 1],
    n_jobs=-1
)


best_model_clf = ImbPipeline([
    ('preprocessor', preprocessor),
    ('smote', SMOTE(
        k_neighbors=5,
        sampling_strategy=0.6912309956335814,
        random_state=42
    )),
    ('voting', voting_clf)
])

best_model_clf.fit(X_train_clf, y_train_clf)
```
